```python
def ad2_spend(low, up ,df):
    clicks=0
    budget=6250000
    win=0
    for i in range(len(df)):
        #print(i)
        y = df['payprice'].iloc[i]
        x = random.sample(range(low, up), 1)
        if budget <= y:
            break
        elif (x>=y and x >=df['slotprice'].iloc[i]):
            #df['win'].iloc[i] = 1
            win+=1
            budget -= df['payprice'].iloc[i]
            clicks += df['click'].iloc[i]
    #df.loc[df["win"]=='','win'] = 0
    spend = 6250000-budget
    #ctr=clicks/(df['win'].sum())
    ctr=clicks/win
    return clicks,spend,ctr, win
```

```python
ad3=validation.copy()
c =[]
sp = []
b = []
pctr = []
low_r =[]
up_r = []
for low in range (15,35):
    for up in range (90,110):
            clicks, spend,ctr, win = ad2_spend(low, up,ad3)
            low_r.append(low)
            up_r.append(up)
            c.append(clicks)
            sp.append(spend)
            b.append(win)
            pctr.append(ctr)
            print(low, up)
            print(clicks,spend,ctr, win)
```
