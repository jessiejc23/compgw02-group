```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#group = pd.read_csv("Group_xx.csv")
train = pd.read_csv("train.csv")
#test = pd.read_csv("test.csv")
#val = pd.read_csv("validation.csv")
```

```python
train.insert(0, 'Imps', 1)
```

```python
## cost spent on clicks (for effective cost)
train['clickcost'] = train['click']*train['payprice']
```

```python
train['slot_size']= train['slotwidth']*train['slotheight']
```

```python
train['OS'],train['browser'] = train['useragent'].str.split('_', 1).str
```

```python
train.head(10)
```

```python
train_df = train.drop(['usertag','slotvisibility','slotformat','bidid','userid','useragent','IP','domain','url','urlid','slotid','creative','keypage'], axis=1)
train_df.head(5)
```

```python
ad = train_df.groupby('advertiser')
```

```python
ad_click_df = pd.DataFrame(ad['click'].sum())
ad_click_df
```

```python
stats = train_df[['Imps','click','bidprice','payprice','advertiser','clickcost']]
stats = stats.groupby('advertiser').sum()
stats
```

```python
stats['CTR %'] = stats['click']/stats['Imps']*100
stats
```

```python
#stats['CPM'] = stats['payprice']/(stats['Imps']/1000)
stats['CPM'] = stats['payprice']/(stats['Imps'])
stats
```

```python
stats['eCPC']=stats['payprice']/stats['click']/1000
stats
```

```python
stats['eCost'] = stats['clickcost']/(stats['payprice'])
stats
```
